```python
kod_yazanlar = {"kerim", "mehmet", "selim"}
kod_yazmayanlar = {"duru", "mehmet", "cem"}
ortak = kod_yazanlar.intersection(kod_yazmayanlar) #iki listedeki ortakları  çeker
print(ortak)
kodyazmayanlar = kod_yazanlar.difference(kod_yazmayanlar) #ilk yazılandaki ikinciden farklı olanları çeker
print(kodyazmayanlar)
imzalistesi = kod_yazmayanlar.union(kod_yazanlar) #iki listediki bütün terimleri toplar
print(imzalistesi,"imza listesi")
```
